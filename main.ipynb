```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.stattools import jarque_bera
from statsmodels.stats.diagnostic import acorr_ljungbox
import openpyxl
import seaborn as sns

data_df = pd.read_excel('USYIELDSLW - 8824.xlsx')

sample_length = np.size(data_df['Month'])
print("sample length =", sample_length)

print(data_df.head())
```

```python
# Get yield columns (assuming they start with 'Y')
yield_cols = [col for col in data_df.columns if col.startswith('Y')]

print(f"Number of yields: {len(yield_cols)}")
print(f"Sample length: {len(data_df)}")

# ==========================================
# PART 1: UNIVARIATE STATISTICS
# ==========================================

# Create a summary statistics table
univariate_stats = pd.DataFrame(index=yield_cols)

for col in yield_cols:
    y = data_df[col].dropna()
    
    univariate_stats.loc[col, 'Mean'] = y.mean()
    univariate_stats.loc[col, 'Std Dev'] = y.std()
    univariate_stats.loc[col, 'Variance'] = y.var()
    univariate_stats.loc[col, 'Skewness'] = stats.skew(y)
    univariate_stats.loc[col, 'Kurtosis'] = stats.kurtosis(y)
    univariate_stats.loc[col, 'Min'] = y.min()
    univariate_stats.loc[col, 'Max'] = y.max()

print("\n" + "="*60)
print("UNIVARIATE STATISTICS")
print("="*60)
display(univariate_stats.round(4))

# ==========================================
# PART 2: AUTOCORRELATIONS
# ==========================================

# Compute autocorrelations at lags 1, 6, 12
acf_results = pd.DataFrame(index=yield_cols)

for col in yield_cols:
    y = data_df[col].dropna()
    
    # Manual autocorrelation calculation
    for lag in [1, 6, 12]:
        if len(y) > lag:
            acf_results.loc[col, f'ACF({lag})'] = y.autocorr(lag=lag)

print("\n" + "="*60)
print("AUTOCORRELATIONS")
print("="*60)
display(acf_results.round(4))

# ==========================================
# PART 3: COVARIANCE AND CORRELATION MATRICES
# ==========================================

# Select only yield columns for covariance/correlation
yields_only = data_df[yield_cols]

# Covariance matrix
cov_matrix = yields_only.cov()
print("\n" + "="*60)
print("COVARIANCE MATRIX")
print("="*60)
display(cov_matrix.round(4))

# Correlation matrix
corr_matrix = yields_only.corr()
print("\n" + "="*60)
print("CORRELATION MATRIX")
print("="*60)
display(corr_matrix.round(4))

# ==========================================
# PART 4: VISUALIZATIONS
# ==========================================

# Plot 1: Time series of all yields
plt.figure(figsize=(12, 6))
for col in yield_cols[::3]:
    plt.plot(data_df.index, data_df[col], label=col, alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Yield (%)')
plt.title('US Government Bond Yields Over Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()

# Plot 2: Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix of Bond Yields')
plt.tight_layout()
plt.show()

selected_yields = [yield_cols[i] for i in [0, 3, 6, 9, 12, 15]]  # Select 6 yields

# Plot 4: ACF plots for selected maturities
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for idx, col in enumerate(selected_yields):
    ax = axes[idx // 3, idx % 3]
    y = data_df[col].dropna()
    
    # Calculate ACF manually
    acf_vals = [y.autocorr(lag=i) for i in range(25)]
    
    ax.stem(range(25), acf_vals, basefmt=' ')
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.axhline(y=1.96/np.sqrt(len(y)), color='red', linestyle='--', linewidth=0.5)
    ax.axhline(y=-1.96/np.sqrt(len(y)), color='red', linestyle='--', linewidth=0.5)
    ax.set_title(f'{col} Autocorrelation')
    ax.set_xlabel('Lag')
    ax.set_ylabel('ACF')

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("ANALYSIS COMPLETE!")
print("="*60)
```

```python

# ===== COVARIANCE MATRIX =====
cov_matrix = yields_only.cov()

# ===== EIGENDECOMPOSITION =====
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort by eigenvalue (descending)
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# ===== COMPUTE PRINCIPAL COMPONENTS =====
# PC_jt = e_j' * R_t (each PC is a linear combination of yields)
yields_demeaned = yields_only - yields_only.mean()
PCs = yields_demeaned @ eigenvectors  # 444 x 17 matrix of PC scores

# Rename columns
PC_names = [f'PC{i+1}' for i in range(len(yield_cols))]
PCs.columns = PC_names

# ===== i. TOTAL VARIANCE EXPLAINED =====
total_var = eigenvalues.sum()
var_explained = eigenvalues / total_var
cumulative_var = np.cumsum(var_explained)

variance_table = pd.DataFrame({
    'Eigenvalue': eigenvalues,
    'Variance Explained': var_explained,
    'Cumulative': cumulative_var
}, index=PC_names)

print("i. TOTAL VARIANCE EXPLAINED BY EACH PC")
display(variance_table.round(7))

# ===== ii. INDIVIDUAL VARIANCE EXPLAINED =====
# Fraction of yield i's variance explained by PC j = (e_ji^2 * lambda_j) / sigma_i^2
individual_var = pd.DataFrame(index=yield_cols, columns=PC_names)

for i, yield_name in enumerate(yield_cols):
    sigma_i_sq = yields_only[yield_name].var()
    for j in range(len(eigenvalues)):
        e_ji = eigenvectors[i, j]
        individual_var.loc[yield_name, PC_names[j]] = (e_ji**2 * eigenvalues[j]) / sigma_i_sq

print("\nii. FRACTION OF INDIVIDUAL VARIANCE EXPLAINED BY EACH PC")
display(individual_var.astype(float).round(4))

# ===== FACTOR LOADINGS (EIGENVECTORS) =====
loadings = pd.DataFrame(eigenvectors, index=yield_cols, columns=PC_names)
print("\nFACTOR LOADINGS (EIGENVECTORS)")
display(loadings.round(4))

# ===== PC SCORES (first few rows) =====
print("\nPRINCIPAL COMPONENTS (first 5 observations)")
display(PCs.head().round(4))
```
